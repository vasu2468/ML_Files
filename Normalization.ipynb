```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
import sklearn
import numpy as np
```

```python
x = np.arange(0,20,0.4)
x
```

```python
a = np.sin(x)
a
```

```python
b = np.cos(x)*50
b
```

```python
Y1 = (a - np.mean(a))/(max(a) - min(a))
Y2 = (b - np.mean(b))/(max(b) - min(b))
```

```python
plt.plot(x,Y1,'blue')
plt.plot(x,Y2,'red')
```
