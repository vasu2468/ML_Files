```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
import sklearn
import numpy as np
```

```python
x = np.arange(0,20,0.4)
x
```

```python
a = np.sin(x)
a
```

```python
b = np.cos(x)*50
b
```

```python
from scipy import stats 
```

```python
IQR1 = stats.iqr(a, interpolation = 'midpoint') 
y1 = (a-np.median(a))/IQR1
```

```python
IQR2 = stats.iqr(b, interpolation = 'midpoint') 
y2 = (b-np.median(b))/IQR2
```

```python
plt.plot(x,y1,'blue')
plt.plot(x,y2,'red')
```
